```python
import pandas as pd
import sqlite3

# Define the path to your CSV file (using raw string to handle backslashes)
csv_file_path = r'C:\Users\shamb\Desktop\International debt analysis\International_debt dataset.csv'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)

# Connect to an in-memory SQLite database
conn = sqlite3.connect(':memory:')

# Load the DataFrame into the SQLite database
df.to_sql('international_debt', conn, index=False, if_exists='replace')

# Create a cursor object to execute SQL queries
cur = conn.cursor()

# SQL query to find the highest amount of repayment, along with country and indicator names
query_highest_repayment = '''
SELECT country_name, indicator_name, MAX(total_repayment) as max_repayment
FROM (
    SELECT country_name, indicator_name, SUM(debt) as total_repayment
    FROM international_debt
    GROUP BY country_name, indicator_name
) AS subquery
GROUP BY country_name, indicator_name
ORDER BY max_repayment DESC
LIMIT 1
'''

# Execute the query
cur.execute(query_highest_repayment)

# Fetch the result
highest_repayment = cur.fetchone()

# Display the result
print("Country with highest repayment:")
print(f"Country Name: {highest_repayment[0]}")
print(f"Indicator Name: {highest_repayment[1]}")
print(f"Highest Repayment Amount: {highest_repayment[2]}")

# Close the connection
conn.close()
```

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Define the path to your CSV file (using raw string to handle backslashes)
csv_file_path = r'C:\Users\shamb\Desktop\International debt analysis\International_debt dataset.csv'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)

# Connect to an in-memory SQLite database
conn = sqlite3.connect(':memory:')

# Load the DataFrame into the SQLite database
df.to_sql('international_debt', conn, index=False, if_exists='replace')

# Create a cursor object to execute SQL queries
cur = conn.cursor()

# SQL query to find the total amount of debt for each country, ordered by debt in descending order
query_debt_by_country = '''
SELECT country_name, SUM(debt) as total_debt
FROM international_debt
GROUP BY country_name
ORDER BY total_debt DESC
'''

# Execute the query
cur.execute(query_debt_by_country)

# Fetch all results
results = cur.fetchall()

# Convert results into a DataFrame for plotting
debt_df = pd.DataFrame(results, columns=['Country', 'Total Debt'])

# Close the connection
conn.close()

# Plotting the data
plt.figure(figsize=(12, 8))
plt.barh(debt_df['Country'], debt_df['Total Debt'], color='skyblue')
plt.xlabel('Total Debt')
plt.ylabel('Country')
plt.title('Total Debt by Country (Highest to Lowest)')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest debt at the top
plt.show()
```

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Define the path to your CSV file (using raw string to handle backslashes)
csv_file_path = r'C:\Users\shamb\Desktop\International debt analysis\International_debt dataset.csv'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)

# Connect to an in-memory SQLite database
conn = sqlite3.connect(':memory:')

# Load the DataFrame into the SQLite database
df.to_sql('international_debt', conn, index=False, if_exists='replace')

# Create a cursor object to execute SQL queries
cur = conn.cursor()

# SQL query to find the total amount of debt by country and indicator
query_debt_by_country_indicator = '''
SELECT country_name, indicator_name, SUM(debt) as total_debt
FROM international_debt
GROUP BY country_name, indicator_name
'''

# Execute the query
cur.execute(query_debt_by_country_indicator)

# Fetch all results
results = cur.fetchall()

# Convert results into a DataFrame for plotting
debt_country_indicator_df = pd.DataFrame(results, columns=['Country', 'Indicator', 'Total Debt'])

# Close the connection
conn.close()

# Pivot the DataFrame to get countries as rows and indicators as columns
pivot_df = debt_country_indicator_df.pivot(index='Country', columns='Indicator', values='Total Debt').fillna(0)

# Plotting the data
plt.figure(figsize=(14, 10))
pivot_df.plot(kind='bar', stacked=True, figsize=(14, 10), colormap='tab20')
plt.xlabel('Country')
plt.ylabel('Total Debt')
plt.title('Comparison of Debt by Indicator for Each Country')
plt.legend(title='Indicator')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```
